```python
import pandas as pd
from pandas import read_csv
from itertools import product
```

```python
df_combination = read_csv(r'D:\2020\MATRIZ_O_D_AUTO\BD\Base_TomTom\Mex Tol 2.csv', encoding='latin-1')
df_combination
```

```python
df_zonvisum = read_csv(r'D:\2020\MATRIZ_O_D_AUTO\BD\ZonasVisum\zv_table_mextol2.csv', encoding='latin-1')
df_zonvisum
```

```python
dict_region_to_ids = {reg: list(ids) for reg, ids in df_zonvisum.groupby("region")["NO"]}
dfs = []
for r1, r2, total in df_combination.itertuples(index=False):
    df = pd.DataFrame(product(dict_region_to_ids[r1], dict_region_to_ids[r2]), 
                      columns=["Origin_id", "Destination_id"])
    df["Total_division"] = total / len(df)
    dfs.append(df)
df = pd.concat(dfs)
df
```

```python
df_final = df.loc[df['Total_division'] != 0 ]
df_final
```

```python
df_final.to_csv(r'D:\2020\MATRIZ_O_D_AUTO\BD\Finales\Matriz_Final_MexTol2.csv')
```
