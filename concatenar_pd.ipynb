```python
import pandas as pd
import geopandas
from pandas import read_csv
from shapely import wkt
from geopandas import GeoDataFrame, points_from_xy
```

```python
#seleccionar tablas
table_1 = read_csv(r'D:\2020\HUIXQUILUCAN\Alertas\alertas_después_model\parte1\FINAL_parte1_.csv', encoding="latin-1")
table_2 = read_csv(r'D:\2020\HUIXQUILUCAN\Alertas\alertas_después_model\parte2\FINAL_parte2.csv', encoding="latin-1")
table_3 = read_csv(r'D:\2020\HUIXQUILUCAN\Alertas\alertas_después_model\parte3\FINAL_parte3.csv', encoding="latin-1")
table_4 = read_csv(r'D:\2020\HUIXQUILUCAN\Alertas\alertas_después_model\parte4\FINAL_parte4.csv', encoding="latin-1")
table_5 = read_csv(r'D:\2020\HUIXQUILUCAN\Alertas\alertas_después_model\parte5\FINAL_parte5.csv', encoding="latin-1")
```

```python
concatenar_tablas = pd.concat([table_1, table_2, table_3, table_4, table_5])
concatenar_tablas.sort_values(by='ts', ascending=True)
```

```python
concatenar_tablas['geoWKT'] = concatenar_tablas['geoWKT'].apply(wkt.loads)

gdf = GeoDataFrame(concatenar_tablas, geometry = 'geoWKT')
gdf 
```
