```python
from pandas.io.json import json_normalize
import pandas as pd
from geopandas import GeoDataFrame
from shapely import wkt
import matplotlib.pyplot as plt
```

```python
geojson = pd.read_json(r'D:\2021\OD_MATRIX_AUTO\OUTPUTS\TomTom_20210204\TJ.json')
```

```python
df = pd.json_normalize(geojson["features"])
df
```

```python
df.dtypes
```

```python
df['region'] = df['properties.name'].astype(str)
df
```

```python
df['geom'] = df['geometry.coordinates'].astype(str)
df
```

```python
df['geom'] = df['geom'].str.replace(", ", " ")
df
```

```python
df['geom'] = df['geom'].str.replace('] [' , ', ', regex=False)
df
```

```python
df['geom'] = df['geom'].str.replace('[[[' , 'POLYGON((', regex=False)
df
```

```python
df['geom'] = df['geom'].str.replace(']]]' , '))', regex=False)
df
```

```python
data = df[['region','geom']]
data
```

```python
data['geom'] = data['geom'].apply(wkt.loads)

gdf = GeoDataFrame(data, geometry = 'geom')
gdf 
```

```python
#Shapefile 1
gdf.to_file(r'D:\2021\OD_MATRIX_AUTO\SHP\TomTom_20210204\TJ.shp', encoding='latin1')
```
